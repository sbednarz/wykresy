```python
import numpy as np
import matplotlib.pyplot as plt


gc_czosnek = np.loadtxt('GC_czosnek_15ml.csv', delimiter=',')
gc_cebula = np.loadtxt('GC_cebula_15ml.csv', delimiter=',')

plt.figure(figsize=(7, 6))
plt.xlim(10,25)
plt.ylim(-0.05,1+0.5)
plt.grid()
plt.plot(gc_czosnek[:,1], gc_czosnek[:,2]+0.5, '-', label='czosnek')
plt.plot(gc_cebula[:,1], gc_cebula[:,2], '-', label='cebula')

plt.xlabel('Czas retencji, min')
plt.ylabel('Sygna')
plt.legend()
plt.savefig('wykres3.png', dpi=300)
```
