```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gc_czosnek = np.loadtxt('GC_czosnek_15ml.csv', delimiter=',')
```

```python
plt.plot(gc_czosnek[:,1], gc_czosnek[:,2], 'b-')
```

```python
plt.xlim(0,35)
plt.ylim(0,3)
plt.plot(gc_czosnek[:,1], gc_czosnek[:,2], 'b-')
```

```python
plt.xlim(6,20)
plt.ylim(0,1)

plt.plot(gc_czosnek[:,1], gc_czosnek[:,2], 'b-')
```

```python
plt.xlim(6,35)
plt.ylim(0,1)
plt.plot(gc_czosnek[:,1], gc_czosnek[:,2], 'b-')
plt.xlabel('Czas retencji, min')
plt.ylabel('Sygnał')
plt.savefig('gc_czosnek1.png', dpi=300)
```
