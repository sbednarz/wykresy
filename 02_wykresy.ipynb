```python
import matplotlib.pyplot as plt
import numpy as np

# PRZYPADEK 2
# rysowanie wykresów z danych liczonych w kodzie

x = np.linspace(0,60)

y1 = 3*x+10
y2 = 0.1*x**2 - 3*x + 10

plt.plot(x,y1,'b-', label='y1')
plt.plot(x,y2,'r--', label='y2')

plt.xlabel('x')
plt.ylabel('y1, y2')

plt.xlim(-100,100) # zakres x
plt.ylim(-50,300) # zakres y

plt.legend(loc='upper left') # loc = location
```

```python
# nakładanie wykresów

# jeden na drugi


import matplotlib.pyplot as plt
import numpy as np

data1=np.loadtxt('hplc.txt') # load data from file to a matrix (array)
data2=np.loadtxt('hplc.txt') # load data from file to a matrix (array)

# seria 1
plt.plot(data1[:,0], data1[:,1], 'k-') # czarna linia
# seria 2
plt.plot(data2[:,0], data2[:,1], 'c--') # niebieska przerywana
```

```python
# nakładanie wykresów

# jeden NAD drugim (stos)


import matplotlib.pyplot as plt
import numpy as np

data1=np.loadtxt('hplc.txt') # load data from file to a matrix (array)
data2=np.loadtxt('hplc.txt') # load data from file to a matrix (array)
data3=np.loadtxt('hplc.txt') # load data from file to a matrix (array)


shifty = 150 # przesunięcie w pionie

# seria 1
plt.plot(data1[:,0], data1[:,1], 'k-') # czarna linia
# seria 2
plt.plot(data2[:,0], data2[:,1] + shifty, 'c--') # niebieska przerywana
# seria 3
plt.plot(data2[:,0], data2[:,1] + 2*shifty, 'g-') # zielona

plt.yticks([]) # wyłączenie skali na osi Y
plt.xlabel('Czas retencji, min')
```

```python
# ZOOM2

import matplotlib.pyplot as plt
import numpy as np

data1=np.loadtxt('hplc.txt') # load data from file to a matrix (array)
data2=np.loadtxt('hplc.txt') # load data from file to a matrix (array)

# seria 1
plt.plot(data1[:,0], data1[:,1], 'k-') # czarna linia
# seria 2
plt.plot(data2[:,0], 3*data2[:,1], 'c--') # niebieska przerywana, 3x wyższy
# seria 3
plt.plot(data1[:,0] + 10, data1[:,1] + 100, 'r-') # czerwona linia, przesunięty w prawo o 10 jednostek i w górę o 100
```
