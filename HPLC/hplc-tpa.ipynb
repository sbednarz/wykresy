```python
import numpy as np
import matplotlib.pyplot as plt
tpa1mg_1 = np.loadtxt('TPA/330_Kwas_Tereftalowy_1mg_ml - MWD 2.1L- Channel 1.txt', delimiter='\t', skiprows=18)
tpa01mg_1 = np.loadtxt('TPA/331_Kwas_Tereftalowy_0p1mg_ml - MWD 2.1L- Channel 1.txt', delimiter='\t', skiprows=18)
tpa001mg_1 = np.loadtxt('TPA/332_Kwas_Tereftalowy_0p01mg_ml - MWD 2.1L- Channel 1.txt', delimiter='\t', skiprows=18)


plt.xlim(6,9)
plt.plot([], [], ' ', label="Stężenie TPA:")
plt.plot(tpa1mg_1[:,0], tpa1mg_1[:,1], color='C0', label='1mg/mL')
plt.plot(tpa01mg_1[:,0], tpa01mg_1[:,1],  color='C1', label='0.1mg/mL')
plt.plot(tpa001mg_1[:,0], tpa001mg_1[:,1],  color='C2', label='0.01mg/mL')
plt.legend()
plt.xlabel("Czas retencji, min")
plt.ylabel("Sygnał")
plt.savefig('wykres_tpa_porownanie.png', dpi=300)
```
