```python
import numpy as np
import matplotlib.pyplot as plt
# https://www.infor.pl/kalkulatory/ilosci_dni.html?fobosId=FOB0000000000000702537

def loa (filename):
    return np.loadtxt(filename, delimiter='\t', skiprows=18)
    

def plo (dataset, label=""):
    plt.plot(dataset[:,0], dataset[:,1], label=label)  


R2_0 = loa('R2/Det2/003_R2_12_02_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_1 = loa('R2/Det2/006_R2_13_02_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_2 = loa('R2/Det2/227_R2_14_02_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_4 = loa('R2/Det2/230_R2_16_02_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_7 = loa('R2/Det2/233_R2_19_02_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_35 = loa('R2/Det2/374_R2_18_03_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_42 = loa('R2/Det2/516_R2_25_03_Spirulina231 - MWD 2.1L- Channel 2.txt')
R2_55 = loa('R2/Det2/569_R2_8_04_Spirulina231 - MWD 2.1L- Channel 2.txt')

TPA01 = loa('TPA/331_Kwas_Tereftalowy_0p1mg_ml - MWD 2.1L- Channel 1.txt')
p021 = loa('pozywka-TPA/Det2/232_TPA_021mg_Brzeczka_czysta - MWD 2.1L- Channel 2.txt')

plt.xlim(5,9)


# uycie funkcji 'plo' mniej kodu:
plo(R2_0, 'R2 - 0 dni')
plo(R2_1, 'R2 - 1 dni')
plo(R2_2, 'R2 - 2 dni')
plo(R2_4, 'R2 - 4 dni')
plo(R2_7, 'R2 - 7 dni')
plo(R2_35, 'R2 - 35 dni')
plo(R2_42, 'R2 - 42 dni')
plo(R2_55, 'R2 - 55 dni')
plo(TPA01, 'TPA 0.1mg woda')
plo(p021, 'TPA 0.21mg + poywka')


plt.xlabel("Czas retencji, min")
plt.ylabel("Sygna")
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='upper left')
```
